```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
X = pd.DataFrame(dataset['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm', 'petal width in cm'])
X.head()
```

```python
y = dataset['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier(max_depth=2)
dectree.fit(X_train,y_train)
```

```python
#Visualize the Decision tree
from sklearn import tree
plt.figure(figsize=(10,5))
tree.plot_tree(dectree,filled=True)
```

```python
y_predict = dectree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('Acc : ', accuracy_score(y_test,y_predict))
print('Conf Matrix : ', confusion_matrix(y_test,y_predict))
print('Classification report :', classification_report(y_test,y_predict))
```

Decision Tree Classifier Pre-pruning and Hyperparameter Tuning

```python
params = {'criterion': ['gini', 'entropy', 'log_loss'],
          'splitter' : ['best', 'random'],
          'max_depth' : [1,2,3,4,5],
          'max_features' : ['auto', 'sqrt', 'log2']}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
grid = GridSearchCV(dectree,param_grid=params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_predict_grid = grid.predict(X_test)
print('Acc : ', accuracy_score(y_test,y_predict_grid))
print('Conf Matrix : ', confusion_matrix(y_test,y_predict_grid))
print('Classification report :', classification_report(y_test,y_predict_grid))
```
