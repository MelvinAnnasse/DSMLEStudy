```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
data
```

```python
print(data['DESCR'])
```

```python
X = pd.DataFrame(data['data'],columns=['age','sex','bmi','bp','tc','ldl','hdl','tch','ltg','glu'])
X.head()
```

```python
y = data['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
X_train.head()
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(8,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
dectreereg = DecisionTreeRegressor()
dectreereg.fit(X_train,y_train)
```

```python
y_predict = dectreereg.predict(X_test)
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
print('r2_score : ',r2_score(y_test,y_predict))
print('mae : ',mean_absolute_error(y_test,y_predict))
print('rmse: ',root_mean_squared_error(y_test,y_predict))
```

```python
#Hyperparameter
params = {'criterion' : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
          'splitter' : ['best', 'random'],
          'max_depth' : [1,2,3,4,5,10,15,20,25],
          'max_features': [1, 'sqrt', 'log2']}
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(dectreereg,params,cv=5,scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_predict_grid = grid.predict(X_test)
print('r2_score : ',r2_score(y_test,y_predict_grid))
print('mae : ',mean_absolute_error(y_test,y_predict_grid))
print('rmse: ',root_mean_squared_error(y_test,y_predict_grid))
```

```python
from sklearn import tree
model = DecisionTreeRegressor(criterion='friedman_mse',max_depth=5,max_features='sqrt',splitter='random')
model.fit(X_train,y_train)
```
